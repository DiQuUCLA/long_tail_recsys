```python
user_feat[2] = user_feat[2].transform(lambda x : -1 if x =='M' else 1)
```

```python
user_feat.fillna(0)
```

```python
item_features = {}
for idx, row in item_feat.iterrows():
    row = row.values
    item_id = row[0]
    row[1] = row[1][1:]
    row[128] = row[128][:-1]
    row[129] = row[129][1:]
    row[256] = row[256][:-1]
    item_features[item_id] = row[1:]
```

```python
user_features = {}
for idx, row in user_feat.iterrows():
    row = row.values
    user_id = row[0]
    user_features[user_id] = row[1:]
```

```python
click_features = []
for idx, c in click_data.iterrows():
    user_id = int(c[1])
    item_id = int(c[0])
    time  = c[2]
    click_feature = [time]
    click_feature.extend(user_features[user_id])
    click_feature.extend(item_features[item_id])
    click_features.append(click_feature)
```
